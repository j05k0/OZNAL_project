**IMPORTS**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
```

```
dataset = pd.read_csv("https://cloudstor.aarnet.edu.au/plus/s/2DhnLGDdEECo4ys/download?path=%2FUNSW-NB15%20-%20CSV%20Files&files=UNSW-NB15_1.csv")
#dataset = pd.read_csv("https://cloudstor.aarnet.edu.au/plus/s/2DhnLGDdEECo4ys/download?path=%2FUNSW-NB15%20-%20CSV%20Files%2Fa%20part%20of%20training%20and%20testing%20set&files=UNSW_NB15_training-set.csv")
dataset.head()
```

**Attacks by sbytes**

```
attacks_by_sbytes = dataset[['Label','proto','sbytes']]
attacks_by_sbytes = attacks_by_sbytes.groupby(['proto','Label']).mean().sort_values(['sbytes'], ascending=False)
attacks_by_sbytes = attacks_by_sbytes.head(15)
attacks_by_sbytes
```

```
plot_attacks_by_sbytes = attacks_by_sbytes.unstack().value_counts.plot(kind='bar')
```
